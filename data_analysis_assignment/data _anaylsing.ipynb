```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
a = pd.read_csv("bmw.csv")
a
```

```python
a.head()
```

```python
a.tail()
```

```python
a.dtypes
```

```python
a["transmission"].replace(["Automatic","Manual","Semi-Auto"],[0,1,2],inplace = True)
a

```

```python
a["fuelType"].replace(["Diesel","Petrol","Other"],[0,1,2],inplace = True)
a
```

```python
a['fuelType'].replace(['Petrol','Diesel','Hybrid','Electric','Other'],[0,1,2,3,4],inplace=True)
a
```

```python
c_dtypes = {'transmission': int,'fuelType': int }
a = a.astype(c_dtypes)
a.dtypes
```

```python
pd.isnull(a).sum().sum()
```

```python
duplicate = a[a.duplicated()]
duplicate
```

```python
b=a.dropna().drop_duplicates()
b
```

```python
b.dtypes
```

```python
X = b.drop(['price','model'], axis=1)
y = b['price']

print('Shape of X = ', X.shape)
print('Shape of y = ', y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=51)

print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
X_test[0, :]
```

```python
lr.predict([X_test[0, :]])
```

```python
lr.predict(X_test)
```

```python
y_test
```

```python
lr.score(X_test, y_test)
```
